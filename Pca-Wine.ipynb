```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
df= pd.read_csv("C:\\Users\\DELL\\Downloads\\wine.csv")
df.head()
```

```python
df.drop(["Type"],axis=1,inplace=True)
df.head()
```

```python
# Converting into numpy array
wine = df.values
wine
```

```python
# Normalizing the numerical data 
wine_normal = scale(wine)
```

```python
wine_normal
```

```python
pca = PCA()
pca_values = pca.fit_transform(wine_normal)
```

```python
pca_values
```

```python
pca_df=pd.DataFrame(pca_values)
pca_df
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
x = pca_values[:,0:1]
y = pca_values[:,1:2]
z = pca_values[:2:3]

# WE GET PC1,PC2 AND PC3 ,NOW PERFORM CLUSTERING ON THIS PC'S
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['pc1','pc2','pc3'])], axis = 1)
finalDf
```

### HIERARCHICAL CLUSTERING

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns
```

```python
dendrogram = sch.dendrogram(sch.linkage(finalDf, method='complete'))

# IF WE CUT AT 8 WE GET 3 CLUSTERS
```

```python
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'complete')
```

```python
y_hc = hc.fit_predict(finalDf)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
finalDf["h_clusters"]= pd.Series(y_hc)
finalDf
```

### K MEANS

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
kmeans_clust=KMeans(n_clusters=2)
kmeans_clust.fit(finalDf)
kmeans_clust.labels_
```

```python
k = list(range(2,8))
TWSS = []
for i in k:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(finalDf)
    WSS = []
    for j in range(i):
        WSS.append(sum(cdist(finalDf.iloc[kmeans.labels_==j,:],kmeans.cluster_centers_[j].reshape(1,finalDf.shape[1]),"euclidean")))
    TWSS.append(sum(WSS))
TWSS
```

```python
plt.plot(k,TWSS,"ro-")
```

### ELBOW CHART SHOWS THE OPTIMUM K VALUE  IS 3

```python
kmeans_clust=KMeans(n_clusters=3)
kmeans_clust.fit(finalDf)
kmeans_clust.labels_
```

```python
finalDf["k_clusters"]=pd.Series(kmeans_clust.labels_)
```
